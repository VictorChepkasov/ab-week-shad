```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df = pd.read_csv('synthetic_gmv_data_1.2.csv')
df
```

```python
data = df.groupby(['user_id', 'group_name'], as_index=False).agg({'gmv': 'sum'})
data
```

```python
a = data.loc[data.group_name == 'control']
b = data.loc[data.group_name == 'test']
```

```python
t = stats.ttest_ind(a=a.gmv, b=b.gmv, equal_var=False)
```

```python
round(t.statistic, 3), round(t.pvalue, 3)
```
